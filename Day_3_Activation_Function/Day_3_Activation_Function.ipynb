## **1. What is an Activation Function?**
An activation function determines the output of a neuron given its input. It decides whether a neuron should be "activated" or not. Without activation functions, a neural network would be equivalent to a linear regression model.

### **Why are Activation Functions Important?**
1. Introduce **non-linearity** to the network.
2. Help the network learn **complex patterns**.
3. Allow stacking multiple layers to form deep architectures.


### **Sigmoid Activation Function**
The Sigmoid function is defined as:

    σ(x) = 1 / (1 + e^(-x))

#### Derivative of Sigmoid:
    σ'(x) = σ(x) * (1 - σ(x))

#### Key Properties:
- Outputs values between 0 and 1.
- Commonly used in binary classification problems.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid
def sigmoid_derivative(x):
    sig = sigmoid(x)
    return sig * (1 - sig)

# Plotting
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(10, 5))
plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, sigmoid_derivative(x), label="Sigmoid Derivative", linestyle='dashed')
plt.title("Sigmoid Function and Its Derivative")
plt.legend()
plt.grid()
plt.show()
```

**Advantages**

Useful for binary classification problems.

Outputs are in the range [0, 1], interpretable as probabilities.

**Disadvantages**

Vanishing gradient problem for large positive or negative inputs.

**Where to Use?**

Last layer of binary classification models.

### **ReLU (Rectified Linear Unit) Activation Function**
The ReLU function is defined as:

    f(x) = x, if x > 0
           0, if x <= 0

#### Derivative of ReLU:
    f'(x) = 1, if x > 0
            0, if x <= 0

#### Key Properties:
- Outputs 0 for negative inputs, helping with sparsity.
- Simple and computationally efficient.


```python
# ReLU Function
def relu(x):
    return np.maximum(0, x)

# Derivative of ReLU
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Plotting
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(10, 5))
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, relu_derivative(x), label="ReLU Derivative", linestyle='dashed')
plt.title("ReLU Function and Its Derivative")
plt.legend()
plt.grid()
plt.show()
```

- Advantages

Computationally efficient.

Reduces the likelihood of the vanishing gradient problem.
- Disadvantages

Can suffer from the "dying ReLU" problem where neurons output 0 for all inputs.

- Where to Use?

Hidden layers in most neural networks.

### **Tanh Activation Function**
The Tanh function is defined as:

    tanh(x) = (e^(x) - e^(-x)) / (e^(x) + e^(-x))

#### Derivative of Tanh:
    tanh'(x) = 1 - (tanh(x))^2

#### Key Properties:
- Outputs values between -1 and 1.
- Zero-centered, making it better for balanced datasets.


```python
# Tanh Function
def tanh(x):
    return np.tanh(x)

# Derivative of Tanh
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Plotting
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(10, 5))
plt.plot(x, tanh(x), label="Tanh")
plt.plot(x, tanh_derivative(x), label="Tanh Derivative", linestyle='dashed')
plt.title("Tanh Function and Its Derivative")
plt.legend()
plt.grid()
plt.show()
```

- Advantages

Outputs are zero-centered, which helps optimization.

Useful for learning complex patterns.

- Disadvantages

Suffers from the vanishing gradient problem.

- Where to Use?

Hidden layers when output values need to range between [-1, 1].

### **Softmax Activation Function**
The Softmax function is defined as:

    softmax(x_i) = e^(x_i) / Σ(e^(x_j)), for j = 1 to n

#### Key Properties:
- Converts logits into probabilities.
- Sum of all outputs equals 1, making it useful for multi-class classification.


```python
# Softmax Function
def softmax(x):
    e_x = np.exp(x - np.max(x))  # Subtracting max for numerical stability
    return e_x / e_x.sum(axis=0)

# Example
logits = np.array([2.0, 1.0, 0.1])
softmax_probs = softmax(logits)
print(f"Logits: {logits}")
print(f"Softmax Probabilities: {softmax_probs}")
```

- Advantages

Outputs are probabilities that sum to 1.

Useful for multi-class classification.

- Disadvantages

Computationally expensive for large number of classes.

- Where to Use?

Last layer of multi-class classification models.

### **3. Comparing Activation Functions**

| **Activation Function** | **Range**         | **Use Case**                      | **Advantages**                   | **Disadvantages**                |
|--------------------------|-------------------|------------------------------------|-----------------------------------|-----------------------------------|
| **Sigmoid**              | (0, 1)           | Binary Classification             | - Outputs probabilities          | - Vanishing gradient problem      |
|                          |                  |                                    | - Smooth and differentiable       | - Outputs not zero-centered       |
| **ReLU (Rectified Linear Unit)** | [0, ∞)   | Hidden Layers                     | - Simple to implement and fast   | - Dying ReLU problem (neurons stuck at 0) |
|                          |                  |                                    | - Helps mitigate vanishing gradients |                                  |
| **Tanh**                 | (-1, 1)          | Hidden Layers                     | - Zero-centered output           | - Vanishing gradient problem      |
|                          |                  |                                    | - Better for balanced data        |                                  |
| **Softmax**              | (0, 1) (sum=1)   | Multi-class Classification         | - Outputs interpretable probabilities | - Computationally expensive for large classes |
|                          |                  |                                    | - Normalizes outputs              |                                  |

### **Key Takeaways**
1. Use **Sigmoid** when you need probabilities (e.g., binary classification).
2. Use **ReLU** in most hidden layers for fast computation and non-linearity.
3. Use **Tanh** when data is zero-centered or when the range (-1, 1) is useful.
4. Use **Softmax** in the output layer for multi-class classification.
